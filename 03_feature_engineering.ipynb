```python
df = pd.read_csv("../outputs/london_cleaned_counts.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values(["count_point_id", "timestamp"]).reset_index(drop=True)

target_col = "all_motor_vehicles"

df["hour_of_day"] = df["timestamp"].dt.hour
df["day_of_week"] = df["timestamp"].dt.dayofweek
df["month"] = df["timestamp"].dt.month
df["year"] = df["timestamp"].dt.year
df["day"] = df["timestamp"].dt.day

df["lag_1"] = df.groupby("count_point_id")[target_col].shift(1)
df["lag_2"] = df.groupby("count_point_id")[target_col].shift(2)
df["lag_3"] = df.groupby("count_point_id")[target_col].shift(3)

df["rolling_mean_3"] = (
    df.groupby("count_point_id")[target_col]
      .shift(1)
      .rolling(3)
      .mean()
      .reset_index(level=0, drop=True)
)

df = df.dropna().reset_index(drop=True)
df.to_csv("../outputs/london_features_dataset.csv", index=False)
print("Saved corrected feature-engineered dataset.")
print(df.head())
print(df.shape)
```
