```python
import pandas as pd

raw_df = pd.read_csv("../raw/dft_traffic_counts_raw_counts.csv")
points_df = pd.read_csv("../raw/count_points.csv")

print("RAW COUNTS HEAD")
print(raw_df.head())

print("\nRAW COUNTS COLUMNS")
print(raw_df.columns.tolist())

print("\nRAW COUNTS SHAPE")
print(raw_df.shape)

print("\nCOUNT POINTS HEAD")
print(points_df.head())

print("\nCOUNT POINTS COLUMNS")
print(points_df.columns.tolist())

print("\nCOUNT POINTS SHAPE")
print(points_df.shape)
```

```python
print("RAW COLUMNS:")
for c in raw_df.columns:
    print(c)

print("\nPOINTS COLUMNS:")
for c in points_df.columns:
    print(c)
```

```python
print("count_point_id in raw_df:", "count_point_id" in raw_df.columns)
print("count_point_id in points_df:", "count_point_id" in points_df.columns)
```

```python
merged_df = raw_df.merge(points_df, on="count_point_id", how="left")

print(merged_df.head())
print(merged_df.shape)
```

```python
merged_df.to_csv("../outputs/merged_raw_counts.csv", index=False)
print("Merged file saved successfully.")
```
